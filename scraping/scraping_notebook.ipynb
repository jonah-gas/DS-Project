```python
import os
from fbref_scraping import Fbref
import pandas as pd
```

```python
# instantiate scraper 
scraper = Fbref()
```

### Fbref Match Statistics

```python
# define scraping lists
season_start_years = [#2017, 2018, 2019, 2020, 2021,
                       2022]
# note: the european big 5 all have yyyy-yyyy seasons string format
# for leagues with yyyy season format we need to adjust the scraping class first
league_ids = [9, 12, 13, 20, 11] # big five european
```

```python
# scraping loop
for league_id in league_ids:
    for ssy in season_start_years:
        scraper.get_league_match_data_df(league_id, ssy, return_df=False, save_csv=True, print_logs=True)
```

### Fbref Wage Data

```python
save_path = r'../data/scraped/fbref/wages/new'
filename = 'team_wages_big_five_2017_to_2022.csv'
# get wage dataframes

df_list = []
for league_id in league_ids:
    for ssy in season_start_years:
        df = scraper.get_wages_df(league_id, ssy, print_logs=True)
        df_list.append(df)
full_df = pd.concat(df_list, axis=0, ignore_index=True)

# save to csv
full_df.to_csv(os.path.join(save_path, filename), sep=';', index=False)
```
